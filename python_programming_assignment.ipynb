```

# Exercise 1: Prime Numbers
# This function checks whether a given number is prime or not.

def is_prime(n):
    """
    Check if a number is prime.
    
    Parameters:
    n (int): The number to check.
    
    Returns:
    bool: True if n is prime, False otherwise.
    """
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example test case
num = 29
output_prime = is_prime(num)
output_prime
```

```

# Exercise 2: Product of Random Numbers
# This program generates two random numbers and checks if the user can guess their product.

import random

def product_quiz():
    """
    Generates two random numbers and asks the user to guess their product.
    Displays whether the answer is correct or not.
    """
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    correct_answer = num1 * num2
    
    print(f"What is the product of {num1} and {num2}?")
    user_answer = int(input("Enter your answer: "))
    
    if user_answer == correct_answer:
        print("Correct!")
    else:
        print(f"Incorrect. The correct answer is {correct_answer}.")

# Call the function
# product_quiz()
```

```

# Exercise 4: Word Counter
# This program counts the number of occurrences of each word in a given text.

from collections import Counter

def word_counter(text):
    """
    Counts the occurrences of each word in a text.
    
    Parameters:
    text (str): The input text to analyze.
    
    Returns:
    dict: A dictionary with words as keys and their counts as values.
    """
    words = text.split()
    word_counts = Counter(words)
    return word_counts

# Example test case
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_counts = word_counter(input_text)

# Print word counts
for word, count in word_counts.items():
    print(f"'{word}': {count}")
```

```

# Exercise 5: Check for Palindrome
# This function checks if a given string is a palindrome.

import string

def is_palindrome(s):
    """
    Check if the input string is a palindrome.
    
    Parameters:
    s (str): The input string to check.
    
    Returns:
    bool: True if s is a palindrome, False otherwise.
    """
    normalized_str = ''.join(char.lower() for char in s if char.isalnum())
    return normalized_str == normalized_str[::-1]

# Example test case
input_string = "racecar"
output_palindrome = is_palindrome(input_string)
output_palindrome
```
